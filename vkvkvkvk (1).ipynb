```python

import numpy as np
from sklearn.metrics import ndcg_score
from collections import defaultdict

def comp_ndcg(data):
    user_predictions = defaultdict(list)
    user_targets = defaultdict(list)
    for index, row in data.iterrows():
        user_predictions[row["user_id"]].append(row["prediction"])
        user_targets[row["user_id"]].append(row["target"])

    # Расчет NDCG для каждого пользователя
    ndcg_values = []
    for user_id in user_predictions.keys():
        #pred,targ =  user_predictions[user_id],list(map(makeint,user_targets[user_id]))
        pred,targ =  user_predictions[user_id],user_targets[user_id]
        if len(targ) == 1:
            print(1)
            if targ[0] == pred[0] and targ[0] == 1:
                ndcg_values.append(1)
            else:
                ndcg_values.append(0)
            continue
        
        ndcg = ndcg_score([targ], [pred])
        #print(user_id,pred,targ,ndcg)
        ndcg_values.append(ndcg)

    # Усреднение значений NDCG
    #print(ndcg_values)
    mean_ndcg = np.mean(ndcg_values)
    print(f"Средний NDCG: {mean_ndcg}")
```

```python
from xgboost import XGBClassifier
```

```python
X_test = test.drop(['feature_4', 'feature_77','target','feature_65','feature_53'], axis=1)
y_test = test['target']
```

```python
model = XGBClassifier(
    max_depth= 4 , learning_rate=0.001, gamma=0.0001, 
    scale_pos_weight =  sum(y == 0) / sum(y == 1),
    subsample=0.8, colsample_bytree=0.8, reg_alpha=0.0005,        
    n_estimators=600, eval_metric = ''
)
model.fit(X, y)
preds = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, preds)

print(f'Значение F1-метрики: {f1}')
```

```python
data = pd.DataFrame({'user_id':test['search_id'],'prediction':preds,'target':y_test})
comp_ndcg(data)
```
